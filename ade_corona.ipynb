# Combine hospital counts with globale strahlung

```python
agg_covid_patients_df
```

```python
forecast["date"] = forecast.ds.dt.date
```

```python
forecast.columns
```

```python
_grp_ = forecast.groupby("date")

globale_strahlung_df = (
    pd.DataFrame({"strglo_yearly": _grp_.yearly.mean(), "strglo_y": _grp_.y.mean()})
    .reset_index()
    .astype({"date": "datetime64[ns]"})
)
```

```python
globale_strahlung_df["rolling_mean_y"] = globale_strahlung_df.strglo_y.rolling(
    14
).mean()
```

```python
globale_strahlung_df.sample(10)
```

```python
combination_df = pd.merge(
    agg_covid_patients_df,
    globale_strahlung_df,
    left_on="date",
    right_on="date",
    how="right",
)
```

```python
combination_df
```

```python
fig = go.Figure()

fig.add_trace(
    go.Scatter(
        x=combination_df.date,
        y=combination_df.ICU_Covid19Patients,
        name="ICU_Covid19Patients",
        opacity=0.5,
    )
)
fig.add_trace(
    go.Scatter(
        x=combination_df.date,
        y=combination_df.strglo_yearly,
        name="strglo_yearly",
        opacity=0.5,
    )
)


fig.add_trace(
    go.Scatter(
        x=combination_df.date,
        y=combination_df.rolling_mean_y - 52,
        name="f(rolling_mean_y)",
        opacity=0.5,
    )
)
# fig.update_layout(barmode="stack")  # relative
fig.show(renderer="browser")
```

```python
file_dir = WORK_DIR / "data" / "bag" / "data" / "COVID19Death_geoRegion_AKL10_w.csv"
```

```python
covid_death_df = pd.read_csv(file_dir).drop(
    columns=[
        "type",
        "type_variant",
        "version",
        "timeframe_all",
        "timeframe_phase2",
        "timeframe_phase2b",
        "timeframe_28d",
        "timeframe_14d",
    ]
)
```

```python
covid_death_ch_df = covid_death_df[
    (covid_death_df.geoRegion == "CH")
    & (covid_death_df.altersklasse_covid19 != "Unbekannt")
]
date_sundays = (covid_death_ch_df.datum_dboardformated + "-7").apply(
    lambda x: datetime.strptime(x, "%G-%V-%u")
)

covid_death_ch_df.loc[:, "date_sundays"] = date_sundays
```

```python
covid_death_ch_df
```

```python
fig = go.Figure()
for (col, data) in covid_death_ch_df.groupby("altersklasse_covid19"):
    fig.add_trace(go.Bar(x=data.date_sundays, y=data.entries, name=col))
fig.update_layout(barmode="stack", xaxis={"categoryorder": "category ascending"})
fig.show(renderer="browser")
```

```python
_grp_ = covid_death_ch_df.groupby("altersklasse_covid19")
pie_of_death_df = pd.DataFrame({"sum_total": _grp_.entries.sum()}).reset_index()
```

```python
pie_of_death_df
```

```python
fig = go.Figure()

fig.add_trace(
    go.Bar(
        y=pie_of_death_df.altersklasse_covid19,
        x=pie_of_death_df.sum_total,
        text=pie_of_death_df.sum_total.astype(str),
        orientation="h",
        opacity=0.5,
    )
)
fig.update_traces(texttemplate="%{text:.2s}", textposition="outside")
fig.show(renderer="browser")
```
